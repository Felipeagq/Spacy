```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
text = 'LA Electronics is looking for new employees in the city of Barranqquilla C.O. to hire them for $ 1 Million pesos and to be able to expand the number of projects in which they can work on 22 of december of 2020'
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
# lo hacemos un objeto de spacy
doc = nlp(text)
```

```python
for token in doc:
    print(f'{token.text:13} : {token.pos_}')
```

```python
from spacy import displacy

displacy.render(doc,
                style='dep',
                options={'distance':100,
                        'compact':True})
```

```python
for ent in doc.ents:
    print(f'{ent.text:18} : {ent.label_}')
```

```python
displacy.render(doc, style='ent')
```
