```python
# importamos la libreria
import spacy
```

```python
# Cargamos un modelo pre-entrenado 
nlp = spacy.load('es_core_news_sm')
```

```python
# Hacemos un documento de prueba
text = 'LA Electronics esta buscando nuevos empleados en la cuidad de Barranqquilla C.O. para contratarlos por $1 Millón pesos y poder ampliar la cantidad de proyecto en los que puede trabajar, entre esos, estan Felipe y Melissa lo cuales empezaron a trabajar en el año 2020'
```

```python
# lo hacemos un objeto de spacy
doc = nlp(text)
```

# Tokenozation

```python
# imprimimos las palabras en doc
for word in doc:
    print(word.text)
```

```python
tokenizer = nlp.tokenizer
```

```python
tokens = tokenizer(text)
len(tokens)
```

# Parts of Speech [POS] Tagging

```python
doc
```

```python
for token in doc:
    print(f'{token.text:13} : {token.pos_}')
```

# Visualizacion

```python
from spacy import displacy
```

```python
displacy.render(doc,
                style='dep',
                options={'distance':100,
                        'compact':True})
```

# Name Entity Recognition [NER]

```python
for ent in doc.ents:
    print(f'{ent.text:18} : {ent.label_}')
```

```python
displacy.render(doc, style='ent')
```
