```python
# importamos la libreria
import spacy
```

```python
# Cargamos un modelo pre-entrenado 
nlp = spacy.load('es_core_news_sm')
```

```python
# Hacemos un documento de prueba
text = 'LA Electronics esta buscando nuevos empleados B-quilla C.O. para contratarlos por $1 Millon y poder ampliar la cantidad de proyecto en los que puede trabajar'
```

```python
# lo hacemos un objeto de spacy
doc = nlp(text)
```

```python
# imprimimos las palabras en doc
for word in doc:
    print(word.text)
```
